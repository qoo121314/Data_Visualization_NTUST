### 補充:如果要保存我們做出來的圖片，有沒有什麼好方法呢?

1. 先把畫布資訊存成變數，用get_figure()方法，來存到 "fig" 這個變數裡面
2. 補上其他的標籤資訊，再對fig使用savefig(),第一個變數放檔案名稱，第二個可以調整解析度(dpi)

```python
sns.boxplot(x='sex' ,y='total_bill', data=tips)
```

```python
#---------1
fig = sns.boxplot(x='sex' ,y='total_bill', data=tips).get_figure()
#---------1
plt.title('Bill vs. Sex',  fontsize=20)

plt.xlabel('Sex', fontsize=20)
plt.ylabel('Total Bill', fontsize=20)

#---------2
fig.savefig("output.png", dpi=800)
#---------2
#完成!
```

1. Pairgrid : 屬量 vs. 屬量
2. Pairplot : Pairgrid的懶人包
3. Facetgrid : 屬質 vs. 屬質

### Pair Grid

```python
iris.head()
```

```python
fig = sns.PairGrid(iris)
```

```python
fig = sns.PairGrid(iris)
fig = fig.map(plt.scatter)
fig
```

```python
fig = sns.PairGrid(iris, hue='species')
fig = fig.map_offdiag(plt.scatter)
fig = fig.map_diag(plt.hist)
fig.add_legend()
#fig.savefig('pair_grid.png', dpi=500)
#保存圖片用上面的方法
```

```python
fig = sns.PairGrid(iris, hue='species', vars=['petal_length','petal_width'])
fig = fig.map_offdiag(plt.scatter)
fig = fig.map_diag(plt.hist, edgecolor='gray')
fig.add_legend()
```

```python
fig = sns.PairGrid(iris, hue='species')
fig = fig.map_diag(plt.hist, edgecolor='gray')

fig = fig.map_upper(sns.scatterplot)
fig = fig.map_lower(sns.kdeplot)

fig.add_legend()
```

## Heat Map and Cluster Map

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flight_mat = flights.pivot('month', 'year', 'passengers')
flight_mat
```

### Heat Map

```python
sns.heatmap(flight_mat)
```

```python
sns.heatmap(flight_mat, cmap='coolwarm', linewidth=1)
```

```python
sns.heatmap(flight_mat, cmap='coolwarm', linewidth=0.5, annot=True)
```

```python
sns.heatmap(flight_mat, cmap='coolwarm', linewidth=0.5, annot=True, fmt ='.0f') #or 'd'
```

```python
# center:指定色彩的中心值

sns.heatmap(flight_mat, cmap='coolwarm', center=flight_mat.loc['March', 1959], \
            annot=True, fmt='d')
```

```python
#藏起顏色條
sns.heatmap(flight_mat, cmap='coolwarm', annot=True, fmt='d', cbar=False)
```
