### 延續99乘法表的例子

```python
#引用套件
import numpy as np
```

```python
# 為了方便接下來的講解 我們改變一下產生的列表的數字 變成11~19的 (9, 1) 矩陣 再乘上1~9的 (1, 9) 矩陣
n = 10
result = np.multiply(
    np.arange(1+10, n+10).reshape(n-1, 1),
    np.arange(1, n).reshape(1, n-1)
)
```

### 方法二 - apply

```python
class_dict={
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
```

```python
df2=df.copy()
```

```python
def change_by_dict(key):
    return class_dict[key]
```

```python
change_by_dict(0)
```

```python
df2 ['target'].apply(change_by_dict )
```

```python
df2 ['target']=df2 ['target'].apply( change_by_dict )
df2.head()
```

#### 法二進階版:匿名函數

```python
df2=df.copy()
```

```python
class_dict={
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
```

```python
df2 ['target']=df2 ['target'].apply( lambda x : class_dict[x] )
df2.head()
```

```python
df2.tail()
```

# 刪除重複值 - drop_duplicates()

```python
df_duplicated=pd.DataFrame(
    [ [1,2],
    [1,5],
    [1,2],
    [1,3],
    [1,3],
    [1,6],
    ], columns=['A', 'B']
)
```

```python
df_duplicated
```

```python
df_duplicated.drop_duplicates() #預設保留第一個值
```

```python
df_duplicated.drop_duplicates(keep= 'last') #保留最後一個
```

### 選取特定欄位

```python
df_duplicated.drop_duplicates('A')  #以A欄位來做重複值刪除
```

```python
df_duplicated.drop_duplicates('B')  #以B欄位來做重複值刪除
```
